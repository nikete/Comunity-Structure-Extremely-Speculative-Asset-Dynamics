```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy
from scipy.stats import kendalltau
```

```python
input_file = "/Users/eaman/research/nikete/Comunity-Structure-Extremely-Speculative-Asset-Dynamics/data/joined_price_network_usd.csv"
data = pd.read_csv(input_file, index_col='symbol')
data['log_severity'] = data['severity_to_average_after_max_volume_weighted'].apply(math.log)
data['magnitude'] = data['normalized_total_volume_before_max'] / data['normalized_total_volume']
data = data.loc[data['magnitude'] != 0]
data['log_magnitude'] = data['magnitude'].apply(math.log)
```

```python
sns.set(style="white")
sns.jointplot(data.log_magnitude, data.log_severity, kind="hex", stat_func=kendalltau)#, color="#4CB391")
plt.savefig('log_magnitude_log_severity.pdf', format='pdf')
```

```python
sns.set(style="white")
sns.jointplot(data.magnitude, data.log_severity, kind="hex", stat_func=kendalltau)#, color="#4CB391")
```
